```python
import numpy as np
import pandas as pd 
from pandas import Series, DataFrame
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2],orient = 'v')
```

```python
sns.boxplot(data=[data1,data2],orient = 'v', whis=np.inf)
```

```python
sns.boxplot(data=[data1,data2],orient = 'h', whis=np.inf)
```

```python
data1 = stats.norm(0,5).rvs(100)

data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])

sns.boxplot(data=[data1,data2],orient = 'v',whis=np.inf)
```

```python
sns.violinplot(data=[data1,data2])
```

```python
sns.violinplot(data=[data1,data2],bw=0.2)
```

```python
sns.violinplot(data=data1,inner='stick')
```
